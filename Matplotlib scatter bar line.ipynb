```python
import matplotlib.pyplot as plt

X = [1,4,6,7,8,9,10]
Y = [2,3,4,5,6,7,9]

plt.figure()
plt.scatter(X, Y, color='red')
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

plt.figure()
plt.bar(X, Y, color='green')
plt.title("Bar Chart")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

plt.figure()
plt.plot(X, Y)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
